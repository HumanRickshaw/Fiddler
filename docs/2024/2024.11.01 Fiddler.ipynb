# What’s the Area of a “Pseudo-Square”?


## [Fiddler](https://thefiddler.substack.com/p/whats-the-area-of-a-pseudo-square)

<p><span>A </span><a href="https://x.com/Anthony_Bonato/status/1843342038730780721" rel>post</a><span> from Anthony Bonato recently caught Tom’s attention. Here’s the image from that post:</span></p>

<p><span>Sure enough, the shape (or one very much like it) has four “sides” of equal length, with four right angles. However, two of these sides are curved (in particular, they are arcs of circles), and two of the right angles are exterior, meaning the </span><em>interior</em><span> angles measure 270 degrees (rather than the usual 90 degrees).</span></p>

<p>Let’s call shapes like this one “pseudo-squares.” A pseudo-square has the following properties:</p>
<ul>
    <li><p>It is a simple, closed curve.</p></li>
    <li><p>It has four sides, all the same length.</p></li>
    <li><p>Each side is either a straight line segment or the arc of a circle.</p></li>
    <li><p>The four sides are joined at four corners, with each corner having an internal angle of 90 degrees or 270 degrees.</p></li>
</ul><p>The pseudo-square pictured above has two straight sides, which run radially between arcs of two concentric circles.</p>

<p><span>Assuming this is a </span><em>unit</em><span> pseudo-square (i.e., each side has length 1), what is its area?</span></p>

In the diagram above:

Knowing the major arc of the small circle is $1$,

<br>
$$\frac{2\pi r}{2\pi} = \frac{1}{2\pi - \phi}$$
<br>
$$\boxed{r = \frac{1}{2\pi - \phi}} $$
<br>

Knowing the minor arc of the large circle is $1$,
<br>
$$\frac{2\pi (r+1)}{2\pi} = \frac{1}{\phi}$$
<br>
$$\boxed{r = \frac{1}{\phi}-1}$$
<br>

Thus,
<br>
$$\frac{1}{2\pi - \phi} = \frac{1}{\phi}-1$$
<br>
$$\phi = (2\pi - \phi) - (2\pi - \phi)\phi$$
<br>
$$0 = \phi^2 - (2\pi + 2)\phi +2pi$$
<br>
$$\phi = \frac{(2\pi + 2) \pm \sqrt{(2\pi + 2)^2 -(4)(1)(2\pi)}}{2}$$
<br>
$$\phi = \frac{(2\pi + 2) \pm \sqrt{4\pi^2 + 8\pi + 4 - 8\pi}}{2}$$
<br>
$$\phi = \boxed{\pi + 1 - \sqrt{\pi^2 + 1} \approx 0.84468}$$
<br>

Solving for r,

$$r \approx \boxed{\frac{1}{0.84468} - 1\approx 0.1839}$$

        <a target = "_blank" href = "2024.11.01fiddler2.png"><img src = "2024.11.01fiddler2.png" style="width:400px"></a>
    </div>
</div>

<ul>
    <li><p style = "color: #39568C;"><strong>Left, Major Arc Sector</strong><p><ul>
            <li>Angle = $2\pi - \phi$</li>
            <li>Radius = $r$</li>
            <li>Area = $\dfrac{2\pi - \phi}{2\pi} \cdot \pi r^2 = \boxed{\pi r^2 - \dfrac{r^2\phi}{2}}$</li>        
        </ul>
    </li><br>    
    <li><p style = "color: #1F968B;"><strong>Right, Minor Arc Sector</strong><p><ul>
            <li>Angle = $\phi$</li>
            <li>Radius = $r+1$</li>
            <li>Area = $\dfrac{\phi}{2\pi} \cdot \pi (r+1)^2 = \boxed{\dfrac{r^2\phi}{2} + r\phi + \frac{\phi}{2}}$</li>
        </ul>
    </li>
</ul>        
<br>
$$A = \left(\pi r^2 - \frac{r^2\phi}{2}\right) + \left(\frac{r^2\phi}{2} + r\phi + \frac{\phi}{2}\right)$$
<br>
$$A = \pi r^2 + \phi\left(r + \frac{1}{2}\right)$$
<br>
$$A \approx \pi \cdot 0.1839^2 + 0.84468(0.1839+0.5)$$

## Answer

$$\boxed{A \approx 0.6839}$$

## Extra Credit

<p>Hopefully, by this point, you have determined the area of a unit pseudo-square with two curved sides and two straight sides.</p>

<p><span>Can you find a unit pseudo-square that has </span><em>three</em><span> curved sides and just one straight side?</span></p>

<p>What is the area of your new unit pseudo-square?</p>

In the diagram above:

Knowing the major arc of the small circle is $1$,

<br>
$$\frac{2\pi r}{2\pi} = \frac{1}{2\pi - 2\phi}$$
<br>
$$\boxed{r = \frac{1}{2\pi - 2\phi}} $$
<br>

Knowing the minor arc of either large circle is $1$,
<br>
$$\frac{2\pi R}{2\pi} = \frac{1}{\phi}$$
<br>
$$\boxed{R = \frac{1}{\phi}}$$
<br>

A third equation relating some combination of $\phi$, $r$, and $R$ is needed, and this comes from $b$.  If the base of the figure were extended left to the center of the right arc, the length of the base can be expressed two ways and set equal.
<br>
$$(R-1) + 1 + (R-1) = R\cos \phi - b + R\cos \phi$$ 
<br>
$$b = 1+ 2R\cos \phi - 2R$$
<br>

Now, using Law of Sines on the triangle,
<br>
$$\frac{\sin(\frac{\pi}{2}-\phi)}{r} = \frac{\sin2\phi}{1+ 2R\cos \phi - 2R}$$
<br>
Cross-multiplying and simplifying with trigonometric identities,
<br>
$$\cos\phi\left(1+ 2R\cos \phi - 2R \right) = 2r\sin\phi\cos\phi$$
<br>
Substituting for $r$ and $R$ from above,
<br>
$$\cos\phi\left(1+ \frac{2\cos \phi - 2}{\phi}\right) = \frac{2\sin\phi\cos\phi}{2\pi - 2\phi}$$
<br>
$$\cos\phi\left(1 + \frac{2\cos \phi - 2}{\phi} - \frac{\sin\phi}{\pi - \phi}\right) = 0**$$
<br>

$\phi = \frac{pi}{2}$ does not make sense, so using [Desmos](https://www.desmos.com/calculator), $\phi \approx 0.7496$.

Solving for $r$,

$$r \approx \boxed{\frac{1}{2\pi - 2\cdot 0.7496} \approx 0.2090}$$

Solving for $R$,

$$R \approx \boxed{\frac{1}{0.7496} \approx 1.3340}$$

The area can be found as follows:
<ul>
    <li><p style = "color: #440154;"><strong>Left & Right Circular Segments</strong><p><ul>
            <li>Angle = $\phi$</li>
            <li>Radius = $R$</li>
            <li>Area = $2 \cdot \frac{1}{2}R^2(\phi - \sin \phi) = R - R^2\sin\phi \approx 0.1215$</li> 
        </ul>
    </li><br>    
    <li><p style = "color: #39568C;"><strong>Trapezoid</strong><p><ul>
            <li>Base<sub>1</sub> = $1$</li>
            <li>Base<sub>2</sub> = $1+ 2R\cos \phi - 2R$</li>
            <li>Height = $R\sin \phi$</li>
            <li>Area = $\dfrac{(2+ 2R\cos \phi - 2R)\cdot R\sin \phi}{2} = (1 + R\cos\phi - R)\cdot R\sin \phi \approx 0.5839$</li> 
        </ul>
    </li><br>    
    <li><p style = "color: #1F968B;"><strong>Triangle</strong><p><ul>
            <li>Base = $1+ 2R\cos \phi - 2R$</li>
            <li>Height = $r\sin (\frac{\pi}{2}-\phi) = r\cos\phi$</li>
            <li>Area = $\dfrac{(1+ 2R\cos \phi - 2R)\cdot r\cos\phi}{2} = \left(\dfrac{1}{2}+ R\cos \phi - R\right)\cdot r\cos\phi \approx 0.0218$</li>
        </ul>
    </li><br>    
    <li><p style = "color: #73D055;"><strong>Top, Major Arc Sector</strong><p><ul>
            <li>Angle = $2\pi - 2\phi$</li>
            <li>Radius = $r$</li>
            <li>Area = $\dfrac{2\pi - 2\phi}{2\pi} \cdot \pi r^2 = \pi r^2 - r^2\phi \approx 0.1045$</li>
        </ul>
    </li>
</ul>        

## Answer

$$A \approx 0.8317$$

**
Note that as $\theta$ gets smaller, the two large unit arcs get further apart, and the top arc is more likely to be a minor arc than a major arc.

In this scenario, $r = \frac{1}{2\phi}$.

The second factor, and thus equation to be solved is now $1 + \frac{2\cos \phi - 2}{\phi} - \frac{\sin\phi}{\phi} = 0$.  
<br>
Continuing the idea that $\theta$ gets smaller,
<br>
$$\lim_{\phi \to 0} \big(1 + \frac{2\cos \phi - 2}{\phi} - \frac{\sin\phi}{\phi}\big)$$
<br>
$$=\lim_{\phi \to 0} 1 + 2\lim_{\phi \to 0} \frac{\cos \phi - 1}{\phi} - \lim_{\phi \to 0} \frac{\sin\phi}{\phi}$$
<br>
$$=1 + 0 - 1 = 0$$
<br>
