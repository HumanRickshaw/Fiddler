# Can You Squeeze the Particles Into the Box?


## [Fiddler](https://thefiddler.substack.com/p/can-you-squeeze-the-particles-into)

<p><span>You have three particles inside a unit square that all repel one another. The energy between each pair of particles is 1/</span><em>r</em><span>, where </span><em>r</em><span> is the distance between them. To be clear, the particles can be anywhere inside the square or on its perimeter. The total energy of the system is the sum of the three pairwise energies among the particles.</span></p>

<p>What is the minimum energy of this system, and what arrangement of the particles produces it?</p>

## Solution

I first explored symmetry and minimizing energy.

<a target = "_blank" href = "2024.12.06-0.png"><img src = "2024.12.06-0.png" style="width:300px"></a>

When placing a third particle given two diametrically opposite particles, is the energy minimized by centering the third point or placing it $x$ off center?  I did not find this to be obvious.

The sum of the energy is :
<br>
$$E(x) = \frac{1}{\sqrt{(a-x)^2+b^2}} + \frac{1}{\sqrt{(a+x)^2+b^2}}$$
<br>
The first derivative of the energy is
$$
\begin{align*}
E'(x) &= -\frac{1}{2}\frac{2(a-x)(-1)}{\left((a-x)^2+b^2\right)^{3/2}} -\frac{1}{2}\frac{2(a+x)}{\left((a+x)^2+b^2\right)^{3/2}} \\
&= \frac{a-x}{\left((a-x)^2+b^2\right)^{3/2}} - \frac{a+x}{\left((a+x)^2+b^2\right)^{3/2}} \\
\end{align*}
$$
<br>
$x = 0$ yields:
<br>
$$\frac{a}{\left(a^2+b^2\right)^{3/2}} - \frac{a}{\left(a^2+b^2\right)^{3/2}} = 0$$

Taking the second derivative of the first fraction of $E(x)$,

$$
\begin{align*}
\frac{d\big[(a-x)\left((a-x)^2+b^2\right)^{-3/2}\big]}{dx} &= (a-x)\left((a-x)^2+b^2\right)^{-5/2}\cdot\left(\frac{-3}{2}\right)(2\cdot(a-x)\cdot-1) - \left((a-x)^2+b^2\right)^{-3/2} \\
&= \left((a-x)^2+b^2\right)^{-5/2}\left(3(a-x)^2 - ((a-x)^2+b^2)\right) \\
&= \frac{2(a-x)^2-b^2}{((a-x)^2+b^2)^{5/2}}\\
\end{align*}
$$
<br>
Taking the second derivative of the second fraction of $E(x)$,
<br>
$$
\begin{align*}
\frac{d\big[(a+x)\left((a+x)^2+b^2\right)^{-3/2}\big]}{dx} &= (a+x)\left((a+x)^2+b^2\right)^{-5/2}\cdot\left(\frac{-3}{2}\right)(2\cdot(a+x)\cdot1) + \left((a+x)^2+b^2\right)^{-3/2} \\
&= \left((a-x)^2+b^2\right)^{-5/2}\left(-3(a+x)^2 + ((a+x)^2+b^2)\right) \\
&= \frac{-2(a+x)^2+b^2}{((a+x)^2+b^2)^{5/2}}\\
\end{align*}
$$
<br>
Thus,
<br>
$$
\begin{align*}
E"(x) &= \frac{2(a-x)^2-b^2}{((a-x)^2+b^2)^{5/2}} - \frac{-2(a+x)^2+b^2}{((a+x)^2+b^2)^{5/2}} \\
&= \frac{2(a-x)^2-b^2}{((a-x)^2+b^2)^{5/2}} + \frac{2(a+x)^2-b^2}{((a+x)^2+b^2)^{5/2}} \\
E"(0) &= \frac{2a^2-b^2}{(a^2+b^2)^{5/2}} + \frac{2a^2-b^2}{(a^2+b^2)^{5/2}} \\
&= \frac{4a^2-2b^2}{(a^2+b^2)^{5/2}} \\
\end{align*}
$$

Symmetry in this scenario yields minimum energy as long as $2a > b\sqrt{2}$.  I shall keep this in mind...



## Solution

I created code that did the following :

<ol>
    <li>Start with :
        <ol>
            <li>A list of particles, initiated with a $[(0,0)]$.</li>
            <li>Initial Energy $=0$ as there are no other particles.</li>
        </ol>
    </li>
    <li>Create a Minimum Energy $=\infty$.</li>
    <li>For each point in the square :
        <ol>
            <li>Pass if the point is already in the list of particles.</li>
            <li>Else, create a temporary energy $=0$.</li>
            <li>For each particle in the particles :
                <ul>
                    <li>Add the energy of the current point and particle to the temp energy.</li>
                </ul>
            </li>
            <li>If the temp energy is less than the Minimum Energy :
                <ul>
                    <li>Save the new point.</li>
                    <li>Save the temp energy as the Minimum Energy.</li>
                </ul>
            </li>
        </ol>
    <li>Return the original particles and the last saved new point.</li>
    <li>Return the sum of the Initial Energy and the saved Minimum Energy.</li>
</ol>

## Answer

I initially thought that induction would be used, that is, the minumum of $p$ particles would be a particle appended to the arrangement of the minumum of $p-1$ particles.  Arrangements with $6-9$ particles quickly debunked that.  I am quite fascinated how close these numbers are for a given number of particles and different approaches to determining the minimum.  I'm quite curious how arrangements look for larger $p$ when there are at least two particles in the interior.
