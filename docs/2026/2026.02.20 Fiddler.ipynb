# Can the Archers Coordinate?


## [Fiddler](https://thefiddler.substack.com/p/can-the-archers-coordinate)

<p>From Bart Wright comes a tale of two archers:</p>

<p><span>Two logicians are trying to earn a fabulous prize as a team. There are three targets, and, to win the prize, each logician must fire a single arrow and hit </span><em>the same</em><span> target as the other. Two of the targets are closer but are otherwise indistinguishable; the logicians know they each have a 98 percent chance of hitting either of these targets. The third target is farther away; the logicians know they each have a 70 percent chance of hitting that target.</span></p>

<p>The logicians can’t cooperate or consult in advance, and they have no knowledge of which target their counterpart is aiming for or whether they are successful.</p>

<p>What is the probability they will win the prize?</p>

## Solution

Both logicians know:

<ol>
    <li>If they both choose any of the three targets, there are 3 winning and 6 losing outcomes.  This yields a maximum of win probability of $\boxed{\dfrac{1}{3}}$.</li>
    <li>If they both choose either of the two indistinguishable targets, there are 2 winning and 2 losing outcomes.  This yields a maximum of win probability of $\boxed{\dfrac{1}{2}}$.</li>
    <li>If they both choose the far away target, there is exactly one winning outcome.  This yields a maximum of win probability of $\boxed{1}$.</li>
</ol>


Setting $x$ as the probability of hitting either of the two indistinguishable targets and $y$ as the farther away target, the general form of winning for each of the above outcomes is:

<ol>
    <li>$\dfrac{2x^2+y^2}{9}$</li>
    <li>$\dfrac{2x^2}{4}$</li>
    <li>$y^2$</li>
</ol>

Comparing the last two,
<br>
\begin{align*}
\dfrac{2x^2}{4} \lt& y^2\\
\dfrac{x}{\sqrt{2}} \lt& y\\
\end{align*}

Note that $\frac{0.98}{\sqrt{2}} \approx 0.693$.

## Answer

Both should aim for the farther away target, yielding a win probability of $$\boxed{0.49}$$
